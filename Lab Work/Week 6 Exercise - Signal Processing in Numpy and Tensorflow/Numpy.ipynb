```python
import numpy as np
import os
import urllib.request
urls_file = '1.txt'  
path = 'img'

file = open(urls_file, 'r')

i = 1
```

```python
try:
    os.mkdir(path)

    for line in file:
        
        try:
            
            iamge_name = path + str(i) + '.jpg'
        
            print(iamge_name, end='\r')
        
            urllib.request.urlretrieve(line, os.path.join(path, iamge_name))
        
            i = i + 1
        
        except:
            
            pass
except:

    
    print("You may need to delete the existing 'img' folder in your directory")

```

```python
import matplotlib.pyplot as plt
import numpy as np

files = [os.path.join('img', file_i)
 for file_i in os.listdir('img')
 if '.jpg' in file_i]


img = plt.imread(files[20])

plt.imshow(img)
```

```python
imgs = [plt.imread(files[file_i])
        for file_i in range(37)]

plt.imshow(imgs[3])
print(imgs[3].shape)
```

```python
import matplotlib.pyplot as plt
import numpy as np


def imgCrop(im, x, y, w, h):
    
    img = im[x:np.clip(x+w,0,im.shape[0]),y:np.clip(y+h,0,im.shape[1])]
    
    return img

for file_i in range(37):
    
    imgs[file_i] = imgCrop(imgs[file_i],0,0,100,100)

plt.imshow(imgs[3])   
print(imgs[3].shape)
```

```python
data = np.array(imgs)
data.shape
print(data.shape)
```

```python
mean_img = np.mean(data, axis=0) 
plt.imshow(mean_img.astype(np.uint8))
print("look at this average person")
```

```python
std_img = np.std(data, axis=0)
plt.imshow(np.mean(std_img, axis=2).astype(np.uint8)) 
print("Mean of all colour channels")
```
